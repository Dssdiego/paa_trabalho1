```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tempo_execucao.csv')
df.head()
```

```python
df.plot()
```

```python
fig, ax = plt.subplots()
sns.barplot(data=df, x='mascara', y='tempo_execucao')
plt.title('Tempo de Execução das Máscaras')
plt.xlabel('Máscara')
plt.ylabel('Tempo de Execução (em ms)')
plt.savefig('tempo_execucao_2.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))
sns.boxplot(data=df, x='maquina', y='tempo_execucao', ax=ax)
plt.title('Tempo de Execução entre Instâncias')
plt.xlabel('Configuração da Instância')
plt.ylabel('Tempo de Execução (em ms)')
plt.savefig('tempo_execucao_1.png')
plt.show()
```
